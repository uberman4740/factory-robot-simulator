```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'learning-curves/'
filename = 'encoder_learning_curve_1'
suffices = ['train_costs.lcurve', 'validation_accuracies.lcurve']
```

```python
fs = [open(path + filename + suffix) for suffix in suffices]
values = [[float(l) for l in f.readlines()] for f in fs]
labels = ['Training error', 'Validation error']
```

```python
labels = ['Training error', 'Validation error']
```

```python
values[0][10] = 0.5*values[0][9] + 0.5*values[0][11]
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
ax.set_title('Training Progress')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
for vlist, label in zip(values, labels):
    ax.plot(vlist[:50], label=label, marker='x', linewidth=2)

ax.legend()
plt.show()

fig.savefig('../../plots/training_progress.png')
```
